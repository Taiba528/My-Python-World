```python
x = df.iloc[:,2:12] 
x.head()
```

```python
#scale all variable between 0 and 1 
from sklearn.preprocessing import MinMaxScaler
scld=MinMaxScaler(feature_range=(0,1))
arr_scld = scld.fit_transform(x)
x_scld=pd.DataFrame(arr_scld,columns=x.columns)
x.head()
x.describe()
x_scld.head()
x_scld.describe()
import warnings
warnings.filterwarnings('ignore')
```

```python
y=df['Base_Sales']
y.head()
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
```

```python
x_scld.head()
```

```python
lm = LinearRegression()
lm = lm.fit(x,y)
coefficients = pd.concat([pd.DataFrame(x.columns),pd.DataFrame(np.transpose(lm.coef_))], axis = 1)
coefficients
lm.intercept_
```

```python
predictions = lm.predict(x)
```

```python
predictions[0:5]
```

```python
#R score of our model
lm.score(x,y)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=1)
train_x.shape
train_y.shape
test_x.shape 
test_y.shape                                              
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(train_x,train_y)
```

```python
predict_test_lm=lm.predict(test_x)
```

```python
print(lm.coef_)
```

```python
df_m=pd.DataFrame({'features':x.columns,'coeff':lm.coef_})
```

```python
df_m
```

```python
df_m=df_m.sort_values(by=['coeff'])
```

```python
df_m.plot(x='features',y='coeff',kind='bar',figsize=(10,10))
```

```python
print("R square value for regression test data is-")
np.round(lm.score(test_x,test_y)*100,2)
```

```python
import numpy as np
from sklearn import metrics
print("simple linear regression Mean square error(MSE) fro Test data:")
np.round(metrics.mean_squared_error(test_y,predict_test_lm),2)
```

```python
from sklearn.metrics import mean_absolute_error
print("simple linear regression Mean absolute error(MAE) fro Test data:")
np.round(metrics.mean_absolute_error(test_y,predict_test_lm),2)
```
