# Build a Multivariate analysis model to determine the effect of marketing spend and ad stock variables on Base sales.

# Create feature & level 

```python
df.head()
```

```python
#scale all variable between 0 and 1 
from sklearn.preprocessing import MinMaxScaler
scld=MinMaxScaler(feature_range=(0,1))
arr_scld = scld.fit_transform(df)
df_scld=pd.DataFrame(arr_scld,columns=df.columns)
df_scld.head()
df_scld.describe()
import warnings
warnings.filterwarnings('ignore')
```

```python
df.columns
```

```python
#creating label & features
x = df_scld.iloc[:, 2:13]
x.head()
```

```python
y = df_scld.iloc[:, 1]
y.head()
```

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm = lm.fit(x,y)
coefficients = pd.concat([pd.DataFrame(x.columns),pd.DataFrame(np.transpose(lm.coef_))], axis = 1)
coefficients
lm.intercept_
```

```python
predictions = lm.predict(x)
```

```python
predictions[0:5]
```

```python
#R² score of our model
lm.score(x,y)
```

# Determine the Ad-stock coefficient for TV spend and Radio spend!

```python
df_m=pd.DataFrame({'features':x.columns,'coeff':lm.coef_})
df_m
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=1)
train_x.shape
train_y.shape
test_x.shape 
test_y.shape                                              
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(train_x,train_y)
```

```python
predict_test_lm=lm.predict(test_x)
```

```python
print(lm.coef_)
```

```python
df_m=pd.DataFrame({'features':x.columns,'coeff':lm.coef_})
```

```python
df_m
```

```python
df_m=df_m.sort_values(by=['coeff'])
```

```python
df_m.plot(x='features',y='coeff',kind='bar',figsize=(10,10))
```

```python
print("R square value for regression test data is-")
np.round(lm.score(test_x,test_y)*100,2)
```

```python
import numpy as np
from sklearn import metrics
print("simple linear regression Mean square error(MSE) fro Test data:")
np.round(metrics.mean_squared_error(test_y,predict_test_lm),2)
```

```python
from sklearn.metrics import mean_absolute_error
print("simple linear regression Mean absolute error(MAE) fro Test data:")
np.round(metrics.mean_absolute_error(test_y,predict_test_lm),2)
```
